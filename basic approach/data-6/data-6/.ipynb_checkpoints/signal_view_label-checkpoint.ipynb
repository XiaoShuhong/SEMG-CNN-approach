```python
# Read log file and do some preprocessing
# input: path - file name
# output: datalst - list of data
#         starts - list of line numbers on all occurance of "StartUp!"
def read_log(path):
    print("Reading file " + path)
    # Read file
    file = open(path)

    # Pre process, drop invalid outputs
    starts = []
    file.seek(0)
    data = file.read()
    datalst = data.split('\n')[2:]

    pre_len = len(datalst)
    i, idx = 0, 0
    while i < len(datalst):
        line = datalst[i]
        if re.match(".*Start", line):
            starts.append(idx)
        #if line.count(',') < 6:
        if line.count(',') != 6:
            datalst.remove(line)
            i -= 1
        i, idx = i + 1, idx + 1
    post_len = len(datalst)

    print("%d invalid data removed." %(pre_len - post_len))
    print("Lines contain 'StartUp!': ", starts)
    print('# Movement samples:', len(starts))
    
    return datalst, starts
```

```python
# Separate data into 6 channels
# input: sample - list of data
# output: channels - list of length 6, each being data for a channel
def extract_channel(sample):
    channels = [[], [], [], [], [], []]
    for line in sample:
        line_data = line.split(',')
        for i in range(6):
            channels[i].append(line_data[i])
    return channels
```

```python
# Extract the middle 1/3 data points of the sample and label as movement
# Extract the first and alst 1/3 data points and label as rest
# input: channels - list of length 6, each being data for a channel
# output: labeled_sample - Nx6 array of the extracted movement interval; 
#                          first channel used as label: 0-rest, 1-fist, 2-open
def label_sample(label, channels):
    sample = np.array(channels, dtype = int).T
    N = sample.shape[0]
    labeled_samples = []
    
    # 1. extract the middle 1/3 data points to label as being in the effective movement interval
    # 2. extract the 10-30% and 70-90% portion of data points to label as rest
    start_portions = [0.1, 0.33, 0.7]
    end_portions   = [0.3, 0.66, 0.9]
    labels         = [0,  label, 0  ]
    for i in range(len(start_portions)):
        start_portion, end_portion, lb = start_portions[i], end_portions[i], labels[i]
        start_idx, end_idx = int(N*start_portion), int(N*end_portion)
        labeled_sample = sample[start_idx:end_idx, :]
        # the first column of the array contains the movement label: 0-rest, 1-fist, 2-open
        labeled_sample[:, 0] = lb
        labeled_samples.append(np.array(labeled_sample))
    return labeled_samples

# 
# input: 
# output: 
def window_labeled_sample(labeled_sample, n_win, n_step):
    labeled_sample_windows = []
    N = labeled_sample.shape[0]
    start_idx, end_idx = 0, n_win
    while end_idx < N:
        labeled_sample_windows.append(labeled_sample[start_idx:end_idx, :])
        start_idx, end_idx = start_idx+n_step, end_idx+n_step
    return np.array(labeled_sample_windows)
```

```python
files = [['rest1.log', 'rest2.log'],  
         ['fist1.log', 'fist2.log'], 
         ['open1.log', 'open3.log']] #用'open2.log'出现了我不认识的bug @Yulin
data_rest, data_fist, data_open = [], [], []

# fs=400Hz  =>  window length = 500ms, increment = 50ms for now
n_win, n_step = 200, 20
all_labeled_sample_windows = []
for label in range(len(files)):
    for file_idx in range(len(files[label])): 
        datalst, starts = read_log(files[label][file_idx])
        # extract individual samples
        for i in range(1,len(starts)-1): # discard the first sample in each file
            start_idx, end_idx = starts[i], starts[i+1]
            sample = datalst[start_idx:end_idx]
            channels = extract_channel(sample)
            # extract movement intervals and windowed segments
            labeled_samples = label_sample(label, channels)
            for labeled_sample in labeled_samples:
                labeled_sample_windows = window_labeled_sample(labeled_sample, n_win, n_step)
                all_labeled_sample_windows.extend(list(labeled_sample_windows))
            
# all_labeled_sample_windows - array: (num_sample_segments, window_length, num_channel)
all_labeled_sample_windows = np.array(all_labeled_sample_windows)
```
