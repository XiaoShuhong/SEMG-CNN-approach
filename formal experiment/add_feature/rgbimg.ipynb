```python
import os
import scipy.io as sio
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
feature_list=['MAV','WL','SSC','ZC','IEMG','VAR','WAMP','RMS','mDWT','HEMG','MAVS','ARC','MNF','PSR','SampEn','CC','HHT']
in_path='C:/Users/11488/Desktop/my semg/formal experiment/signal_with_feature'
image_path='C:/Users/11488/Desktop/my semg/formal experiment/image_with_feature'
```

```python
for s in range(1,28):
    for fea in feature_list:
        for label in ['train','test'] :
            image_f=os.path.join(image_path,
                  's'+str(s),
                    fea,
                    label
                    )
            if os.path.isdir(image_f) is False:
                os.makedirs(image_f)
```

```python
def get_img(s,fea):
    for g in range(1,53):
        p=os.path.join(in_path,
                     's'+str(s),
                       'g'+str(g)
                      )
        mat=sio.loadmat(p)
        for t in ['tr1','tr4','tr9']:
            target=mat[str(fea)+t]
            path=os.path.join(image_path,
                     's'+str(s),
                       str(fea),
                       'test',
                        's'+str(s)+'g'+str(g)+t+'.png'
                              
                      )
            Draw_RGBimg(target,path)
        for t in ['tr0','tr2','tr3','tr5','tr6','tr7','tr8']:
            target=mat[str(fea)+t]
            path=os.path.join(image_path,
                     's'+str(s),
                       str(fea),
                       'train',
                        's'+str(s)+'g'+str(g)+t+'.png'
                              
                      )
            Draw_RGBimg(target,path)
            
            
                     
            
    
    
```

```python
from PIL import Image
def Draw_RGBimg(signal,path):
    mini=signal.min()
    if mini<0:
        signal+=abs(mini)
    if signal.max()!=0:
        signal=signal/signal.max()
        signal*=255
    im = Image.fromarray(signal)
    im = im.convert('RGB') 
    im.save(path)
```

```python
for s in range(26,28):
    for fea in feature_list:
        get_img(s,fea)
```
