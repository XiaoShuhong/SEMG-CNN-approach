```python
def Draw_Gesture(mat,subject):
    ges=mat['gesture'][0,0]
    if (ges<1 or ges>52):
        print('wrong gesture read')
        return 
    else:
        plt.figure(figsize=(40,5))
        for ch in range(10):
            key='gesture_subject'+str(subject)+'_channel'+str(ch)
            signal=mat[key]
            plt.plot(signal[0,:],label='channel'+str(ch+1))
            plt.title('signal of gesture'+str(ges)+'for subject'+str(subject))
            plt.xlabel('n (sample point index)')
            plt.ylabel('Intensity')
            plt.legend(loc="upper right")
        return

    
    
```

```python
def Low_Pass_Filter(signal,cut,fs,order,zero_phase=False):
    nyq = 0.5 * fs
    cut = cut / nyq
    b,a= butter(order, cut, btype='low')
    y = (filtfilt if zero_phase else lfilter)(b, a, signal)
    return y

    
```

```python
def Split_Trials(signal,onset):
    res=[[] for _ in range(TRIALS)]
    head=0
    tail=0
    for i in range(TRIALS):
        head=tail
        tail=tail+int(onset[i])
        res[i]=signal[head:tail]
    return res
```

```python
def Window_Sliding(signal,window_num,window_size):
    step=(len(signal)-window_size)//(window_num-1)
    res=[]
    for i in range(window_num):
        start=i*step
        end=start+window_size
        res.append(signal[start:end])
    return res
        
    
    
```

```python
def Extract_Feature(signal,feature):
    func='emg_features.'+feature
    tot_res=[]
    for ch in range(signal.shape[0]):
        res=[]
        for win in range(signal.shape[1]):
            res.append(eval(str(func))(signal[ch,win,:]))
        tot_res.append(res)
    return tot_res
    
    
    
    
```
